```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# NUMPY
arr = np.array([10, 20, 30, 40, 50])

print("Array:")
print(arr)

print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))

# PANDAS
data = {
    "Name": ["A", "B", "C"],
    "Marks": [80, 90, 95]
}

df = pd.DataFrame(data)

print(df)

# MATPLOTLIB
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.show()
```
